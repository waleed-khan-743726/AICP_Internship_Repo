Q.7: Draw a horizontal bar graph for items in each category against the calories.
Here is one sample for all items in Beef & Pork, you have to do it for items in each category. Also, write
your observation.


```python
categories = df['Category'].unique()

for category in categories:
    # Filter DataFrame for items in the current category
    category_df = df[df['Category'] == category]
    # Sort DataFrame by calories in descending order
    category_df = category_df.sort_values(by='Calories', ascending=False)
    
    # Plotting graph
    plt.figure(figsize=(20, 20))
    plt.barh(category_df['Item'], category_df['Calories'], color='skyblue')
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.title(f'Calories in {category}')
    plt.gca().invert_yaxis()  # Invert y-axis to have the item with the highest calories at the top
    plt.show()
```

```python
import plotly.graph_objs as go
import plotly.express as px
for category in categories:
    category_df = df[df['Category'] == category]
    category_df = category_df.sort_values(by='Calories', ascending=False)

    fig = go.Figure()
    fig.add_trace(go.Bar(
        x=category_df['Calories'],
        y=category_df['Item'],
        orientation='h',
        marker=dict(color=px.colors.qualitative.Plotly[0])  # Using different colors for each category
    ))

    fig.update_layout(
        title=f'Calories in {category}',
        xaxis_title='Calories',
        yaxis_title='Item',
        plot_bgcolor='rgba(0,0,0,0)'
    )

    fig.show()
```

Observation:<br>
Each horizontal bar graph is interactive, allowing users to hover over bars for specific information.<br>
Different colors are used for each category to enhance visual appeal and make it easier to distinguish between categories.<br>
